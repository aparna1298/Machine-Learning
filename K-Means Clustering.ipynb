```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df=pd.read_csv('./Mall_Customers.csv')
```

```python
df
```

```python
x=df.iloc[:,[3,4]].values
#here we dont need y, instead we need to predict the cluster for future data
```

```python
x
```

USING THE ELBOW METHOD TO FIND THE OPTIMAL NUMBER OF CLUSTERS


```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=0) #kmeans++ is for the next cluster
    kmeans.fit(x)
    wcss.append(kmeans.inertia_) #inertia is the method to calculate the wcss and then append it into the list
wcss
    
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xlim(1,10) #x values
plt.show()
```

```python
#k-means++ -> selects initial cluster centres for k-means clustering in a smart way to speed up convergence
```

```python
#inertia_ :(float) Sum of squared distances of samples to their closest cluster center.
#The main difference is that when you say , ms.fit(x), x is your labeled/train dataset , i.e. you are predicting 
#on an unlabeled dataset with fit_predict
```

```python
#Fitting k_means to the dataset
kmeans=KMeans(n_clusters=5, init='k-means++', random_state=0)
y_kmeans=kmeans.fit_predict(x)
```

```python
y_kmeans #tells the cluster that data point belongs too
```

```python
plt.scatter(x[y_kmeans ==0,0], x[y_kmeans==0,1],s=100,c="red",label="Cluster1")
plt.scatter(x[y_kmeans ==1,0], x[y_kmeans==1,1],s=100,c="blue",label="Cluster2")
plt.scatter(x[y_kmeans ==2,0], x[y_kmeans==2,1],s=100,c="green",label="Cluster3")
plt.scatter(x[y_kmeans ==3,0], x[y_kmeans==3,1],s=100,c="orange",label="Cluster4")
plt.scatter(x[y_kmeans ==4,0], x[y_kmeans==4,1],s=100,c="black",label="Cluster5")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],s=100,c="yellow",label="centroid")

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show() #s is for size of the points
```
